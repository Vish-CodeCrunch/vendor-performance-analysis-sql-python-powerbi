## Aggregate data from different table to create final summary table

```python
purchases.columns
```

```python
inventory.columns
```

```python
purchase_inventory = pd.read_sql_query(""" 
    SELECT
    po.vendor_id,
    po.product_id,
    COUNT(*) AS total_orders,
    SUM(po.quantity_ordered) AS total_units_purchased,
    SUM(inv.quantity_received) AS total_units_received,
    SUM(inv.damaged_units) AS total_units_damaged,
    SUM(inv.accepted_units) AS total_units_accepted,
    SUM(po.purchase_amount) AS total_purchase_amount_raw,
    AVG(po.purchase_amount) AS avg_purchase_amount_raw,
    SUM(CASE WHEN po.status = 'Delivered' THEN 1 ELSE 0 END) AS purchase_status_delivered,
    ROUND(SUM(CASE WHEN po.status = 'Pending' THEN 1 ELSE 0 END),0) AS purchase_status_pending,
    ROUND(SUM(CASE WHEN po.status = 'Canceled' THEN 1 ELSE 0 END)) AS purchase_status_cancelled,
    ROUND(AVG(DATEDIFF(inv.receive_date, po.delivery_due_date)),1) AS avg_delivery_days_delay_raw,
    ROUND(100.0 * SUM(CASE WHEN inv.receive_date <= po.delivery_due_date THEN 1 ELSE 0 END) / COUNT(*), 2) AS fulfillment_rate_raw
FROM purchases AS po
JOIN inventory AS inv ON po.purchase_id = inv.purchase_id
GROUP BY po.vendor_id, po.product_id;
""",connection)

purchase_inventory
```

```python
sales.columns
```

```python
sales_agg = pd.read_sql_query("""
    SELECT
        vendor_id,
        product_id,
        SUM(quantity_sold) AS total_sales_quantity,
        SUM(total_sale_value) AS total_sales_revenue_raw,
        SUM(returned_units) AS total_units_returned
    FROM sales
    GROUP BY vendor_id, product_id
    """,connection)

sales_agg
```

```python
vendor_invoices.columns
```

```python
invoice_agg = pd.read_sql_query(""" 
    SELECT
        vendor_id,
        COUNT(*) AS invoice_count,
        SUM(invoice_amount) AS total_invoice_amount_raw,
        AVG(invoice_amount) AS avg_invoice_amount_raw,
        AVG(discount_given) AS avg_discount_given_raw,
        AVG(late_payment_penalty) AS avg_late_penalty_raw
    FROM vendor_invoices
    GROUP BY vendor_id
""",connection)

invoice_agg
```

## Combining All query under one query to make vendors_sales_summary table

```python
import time

START_TIME = time.time()

vendors_sales_summary = pd.read_sql_query("""

WITH 
    purchase_inventory AS (
        SELECT
            po.vendor_id,
            po.product_id,
            COUNT(*) AS total_orders,
            SUM(po.quantity_ordered) AS total_units_purchased,
            SUM(inv.quantity_received) AS total_units_received,
            SUM(inv.damaged_units) AS total_units_damaged,
            SUM(inv.accepted_units) AS total_units_accepted,
            SUM(po.purchase_amount) AS total_purchase_amount_raw,
            AVG(po.purchase_amount) AS avg_purchase_amount_raw,
            SUM(CASE WHEN po.status = 'Delivered' THEN 1 ELSE 0 END) AS purchase_status_delivered,
            SUM(CASE WHEN po.status = 'Pending' THEN 1 ELSE 0 END) AS purchase_status_pending,
            SUM(CASE WHEN po.status = 'Canceled' THEN 1 ELSE 0 END) AS purchase_status_canceled,
            ROUND(AVG(DATEDIFF(inv.receive_date, po.delivery_due_date)),1) AS avg_delivery_delay_days_raw,
            ROUND(100.0 * SUM(CASE WHEN inv.receive_date <= po.delivery_due_date THEN 1 ELSE 0 END) / COUNT(*), 2) AS fulfillment_rate_raw
        FROM purchases AS po
        JOIN inventory AS inv ON po.purchase_id = inv.purchase_id
        GROUP BY po.vendor_id, po.product_id
    ),

    sales_agg AS (
        SELECT
            vendor_id,
            product_id,
            SUM(quantity_sold) AS total_sales_quantity,
            SUM(total_sale_value) AS total_sales_revenue_raw,
            SUM(returned_units) AS total_units_returned
        FROM sales
        GROUP BY vendor_id, product_id
    ),

    invoice_agg AS ( 
        SELECT
            vendor_id,
            COUNT(*) AS invoice_count,
            SUM(invoice_amount) AS total_invoice_amount_raw,
            AVG(invoice_amount) AS avg_invoice_amount_raw,
            AVG(discount_given) AS avg_discount_given_raw,
            AVG(late_payment_penalty) AS avg_late_penalty_raw
        FROM vendor_invoices
        GROUP BY vendor_id
    )

SELECT
    v.vendor_id,
    v.vendor_name,
    v.vendor_rating,
    v.active_status,
    p.product_id,
    p.product_name,
    p.category,

    pi.total_orders,
    pi.total_units_purchased,
    ROUND(pi.total_purchase_amount_raw, 2) AS total_purchase_amount,
    ROUND(pi.avg_purchase_amount_raw, 2) AS avg_purchase_amount,
    pi.purchase_status_delivered,
    pi.purchase_status_pending,
    pi.purchase_status_canceled,

    pi.total_units_received,
    pi.total_units_damaged,
    pi.total_units_accepted,
    ROUND(pi.avg_delivery_delay_days_raw, 2) AS avg_delivery_delay_days,
    ROUND(pi.fulfillment_rate_raw, 2) AS fulfillment_rate,

    s.total_sales_quantity,
    ROUND(s.total_sales_revenue_raw, 2) AS total_sales_revenue,
    s.total_units_returned,

    i.invoice_count,
    ROUND(i.total_invoice_amount_raw, 2) AS total_invoice_amount,
    ROUND(i.avg_invoice_amount_raw, 2) AS avg_invoice_amount,
    ROUND(i.avg_discount_given_raw, 2) AS avg_discount_given,
    ROUND(i.avg_late_penalty_raw, 2) AS avg_late_penalty

FROM purchase_inventory pi
JOIN vendors v ON pi.vendor_id = v.vendor_id
JOIN products p ON pi.product_id = p.product_id
LEFT JOIN sales_agg s ON pi.vendor_id = s.vendor_id AND pi.product_id = s.product_id
LEFT JOIN invoice_agg i ON pi.vendor_id = i.vendor_id

""", connection)

END_TIME = time.time()

print("✅ Summary table generated in", round(END_TIME - START_TIME, 2), "seconds")
vendors_sales_summary
```

This query generates a vendor-wise sales and purchase summary, which is valuable for:

****Performance Optimization:****

* The query involves heavy joins and aggregations on large datasets like sales and purchases.

* Storing the pre-aggregated results avoids repeated expensive computations.

* Helps in analyzing sales, purchases, and pricing for different vendors and brands.

* Future Benefits of Storing this data for faster Dashboarding & Reporting.

* Instead of running expensive queries each time, dashboards can fetch data quickly from vendor_sales_summary.

### Adding Some Useful Columns for analysis

```python
vendors_sales_summary['Gross_Profit'] = vendors_sales_summary['total_sales_revenue'] - vendors_sales_summary['total_purchase_amount']
```

```python
vendors_sales_summary['Profit_Margin'] = (vendors_sales_summary['Gross_Profit'] / vendors_sales_summary['total_sales_revenue'])*100
```

```python
vendors_sales_summary['Stock_Turnover'] = vendors_sales_summary['total_sales_quantity'] / vendors_sales_summary['total_units_purchased']
```

```python
vendors_sales_summary['Sales_Purchase_Ratio'] = vendors_sales_summary['total_sales_revenue'] / vendors_sales_summary['total_purchase_amount']
```

```python
vendors_sales_summary[['Sales_Purchase_Ratio','Gross_Profit','Profit_Margin','Stock_Turnover']] = vendors_sales_summary[['Sales_Purchase_Ratio','Gross_Profit','Profit_Margin','Stock_Turnover']].round(2)
```

### Creating vendors_sales_summary table in database

```python
with engine.connect() as connection:
    result = connection.execute(text("""
    CREATE TABLE vendors_sales_summary (
    vendor_id INT,
    vendor_name VARCHAR(255),
    vendor_rating DECIMAL(15,2),
    active_status VARCHAR(50),
    product_id INT,
    product_name VARCHAR(255),
    category VARCHAR(100),
    total_orders INT,
    total_units_purchased INT,
    total_purchase_amount DECIMAL(15,2),
    avg_purchase_amount DECIMAL(15,2),
    purchase_status_delivered INT,
    purchase_status_pending INT,
    purchase_status_canceled INT,
    total_units_received INT,
    total_units_damaged INT,
    total_units_accepted INT,
    avg_delivery_delay_days FLOAT,
    fulfillment_rate DECIMAL(15,2),
    total_sales_quantity INT,
    total_sales_revenue DECIMAL(15,2),
    total_units_returned INT,
    invoice_count INT,
    total_invoice_amount DECIMAL(15,2),
    avg_invoice_amount DECIMAL(15,2),
    avg_discount_given DECIMAL(10,2),
    avg_late_penalty DECIMAL(7,2),
    Gross_Profit DECIMAL(15,2),
    Profit_Margin DECIMAL(15,2),
    Stock_Turnover DECIMAL(5,2),
    Sales_Purchase_Ratio DECIMAL(15,2)
);
    
"""))

result
```
